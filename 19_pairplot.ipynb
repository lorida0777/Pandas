```python
titanic = pd.read_excel('titanic3.xls')
titanic.head()
```

```python
titanic = titanic.drop(['name',  'ticket', 'cabin', 'boat', 'body', 'home.dest','embarked'], axis=1, errors='ignore')
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='survived', y='age', data=titanic, hue='sex')
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.distplot(titanic['fare'])
```
